# Image Processing

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, color
```

```python
# Load image using Matplotlib
fruits = plt.imread('fruits-2.jpg')
```

```python
# Show image
plt.imshow(fruits)
plt.show()
```

```python
type(fruits)
```

```python
# Obtaining the red values of the image
red = fruits[:,:,0]

# Obtaining the green values of the image
green = fruits[:,:,1]

# Obtaining the blue values of the image
blue = fruits[:,:,2]
```

```python
# Plotting the red image
plt.imshow(red, cmap='gray')
plt.title('Red')
plt.show()
```

```python
# Plotting the green image
plt.imshow(green, cmap='gray')
plt.title('Green')
plt.show()
```

```python
# Plotting the blue image
plt.imshow(blue, cmap='gray')
plt.title('Blue')
plt.show()
```

```python
# Shape of the image
fruits.shape
```

```python
#Size of the image
fruits.size
```

```python
# Flip the image in up direction
vertical = np.flipud(fruits)

# Flip the image in left direction
horizontal = np.fliplr(fruits)
```

```python
# Show the images
plt.imshow(vertical)
plt.title('Vertical')
plt.show()

plt.imshow(horizontal)
plt.title('Horizontal')
plt.show()
```

```python
# histogram of the red image
plt.hist(red.ravel(), bins=256)
plt.title('Red')
plt.show()
```

```python
# histogram of the green image
plt.hist(green.ravel(), bins=256)
plt.title('Green')
plt.show()
```

```python
# histogram of the blue image
plt.hist(blue.ravel(), bins=256)
plt.title('Blue')
plt.show()
```

```python
# Load 4.1.01 image
image = plt.imread('4.1.01.tiff')
```

```python
# Show the image
plt.imshow(image)
plt.show()
```

```python
# Obtain the red values of the image
red = image[:,:,0]
```

```python
# Plot the red image with bins = 256
plt.hist(red.ravel(), bins=256)
plt.title('Red')
plt.show()
```

```python
# Obtain the optimal threshold value
thresh = 127

# Obtain the binary image
binary = image > thresh

# Plot the original and binary images
plt.imshow(image)
```

```python
def show_image(image, title='Image', cmap_type='gray'):     
    plt.imshow(image, cmap=cmap_type)         
    plt.title(title)     
    plt.axis('off')     
    plt.show()


show_image(binary)
```
